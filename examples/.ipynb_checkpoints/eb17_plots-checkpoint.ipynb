# Using redshift

```python
M = np.linspace(-15, -24.5, 100)
```

```python
zs = [0.1, 0.2, 0.3, 0.5, 0.7, 0.9, 1.1]
alphas = [-1.05, -1.175, -1.3, -1.3, -1.3, -1.3, -1.3]
phi_stars = [14.9*0.7**3, 4.32, 3.53, 3.23, 3.46, 3.78, 2.5] # 10^-3 h^3 Mpc^-3 mag^-1  (h = 1)
Mb_stars = [-20.44, -20.54, -20.64, -20.97, -21.08, -21.22, -21.39]  # + 5 logh70
```

```python
phis = []
for i, z in enumerate(zs):
    phis.append(schechter_fn(M, phi_stars[i], Mb_stars[i], alphas[i]))
```

```python
plt.figure(figsize=(10,8))
for i in range(len(zs)):
    l = plt.plot(M, phis[i]*10**(-3), label=f'z_mid = {zs[i]}')#, s=5)#, marker='+')
    plt.axvline(x=Mb_stars[i] + 5, color = l[0].get_c(), alpha = 0.7, linestyle = '--')
plt.axvline(x=-24, color = 'k', alpha = 0.7, linestyle = '--')
plt.yscale('log')
plt.xlabel('M', fontsize=15)
plt.ylabel(r'$\phi [Mpc^{-3} mag^{-1}]$', fontsize=15)
plt.legend()
plt.xlim([-15, -25])
plt.ylim([10**(-8), 10**(-1)])
plt.grid()
```

```python
# z_s = np.linspace(0.1, 1.1, 11)
z_s = [0.3, 0.5, 0.9, 1.1]
```

```python
plt.figure(figsize=(10,8))
for i, z in enumerate(z_s):
    if z < 1.2:
        rs = []
        ns = []
        ps = []
        for R_frb in np.linspace(1,1000,100):
            p, n = p_eb17_z(z, R_frb, ret_numgal = True)
            ns.append(n)
            rs.append(R_frb)
            ps.append(p)
        plt.plot(np.log10(rs), np.log10(ns), label=f'z < {z:.2f}')
plt.xlabel(r'R_FRB (arcsec)', fontsize=15)
plt.ylabel(r'$\Sigma(\leq M).f_A.V_C(\leq z)$', fontsize=15)
plt.legend()
plt.grid()
```
